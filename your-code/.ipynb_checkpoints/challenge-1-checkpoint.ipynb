#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

figura,tabla = plt.subplots(2)

tabla[0].plot(x,y, color = "green", linewidth = 3)
tabla[1].plot(x,z, color = "orange", linestyle = "dashed")
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)

plt.subplot(1,2,2)
plt.plot(x,y,linewidth=4,color="purple", linestyle='dashed')

plt.subplot(1,2,1)
plt.plot(x,z,lw=5, color="red",ls='dotted')


#estructura = plit.subplot(nrows, ncolumns, index)

#nrows = cuántas gráficas quieres por fila
#ncolumns = cuántas gráficas quieres acomodadas en columna 
#index = dicta el orden en el que van las gráficas, si es igual en ambas, las junta en una 
#sola gráfica
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here

figura,tabla = plt.subplots(2, figsize =(15,10))

tabla[0].plot(x,y, color = "blue", linewidth = 3, linestyle = "dashdot")
tabla[1].plot(x,z, color = "yellow", linewidth = 7, linestyle = "dashed")

tabla[0].set_xlabel('hola')
tabla[1].set_xlabel('bye')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
figura, axes = plt.subplots(1, 2, figsize=(12,6))

axes[0].plot(x, x**2,x, np.exp(x)) #estamos definiendo la línea azul y naranja 
#pasamos "x" y "y" para cada una
axes[0].set_title("Escala Normal")

axes[1].plot(x, x**2,x, np.exp(x)) #np.exp(x) regresa "e" elevado a la "x"
axes[1].set_xscale("log") #definimos la escala que queremos que vaya en el eje de las "x"
axes[1].set_yscale("log") #definimos la escala que queremos que vaya en el eje de las "y"
axes[1].set_title("Escala logarítmica")
```
